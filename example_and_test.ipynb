```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from find_f import calc
```

```python
def kauf_line():
    x=np.linspace(-2, 0, 100)
    y = (0.61/(x - 0.05) + 1.3)
    return x, y

def kewl_line():
    x=np.linspace(-2, 0.4, 100)
    y = (0.61/(x - 0.47) + 1.19)
    return x, y
```

```python
x_points = np.linspace(-2, 0.5, 100)
y_points = np.linspace(-1.5, 1.5, 100)

X, Y = np.meshgrid(x_points, y_points)

mean_values = np.zeros_like(X)

for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        x_val = X[i, j]
        y_val = Y[i, j]
        result = calc(x_val, y_val)
        mean_values[i, j] = result['f_mean']
```

```python
# Plotting
plt.figure(figsize=(8, 6))

# Plot Kaufmann and Kewley lines
plt.plot(kauf_line()[0], kauf_line()[1], c='r')
plt.plot(kewl_line()[0], kewl_line()[1], c='r')

# Scatter the points based on the condition
plt.scatter(X, Y, c=mean_values)
plt.colorbar(label=r'Fractional AGN Contribution')
plt.text(-1.5, -0.5, '0% AGN', c='w')
plt.text(0.05, 1, '100% AGN', c='k')
plt.text(-0.2, -1., '50% AGN', c='w')

plt.xlim(-2, 0.5)
plt.ylim(-1.25, 1.25)
plt.xlabel(r'[NII] / H$\alpha$')
plt.ylabel(r'[OIII] / H$\beta$')
#plt.legend()
plt.show()
```
